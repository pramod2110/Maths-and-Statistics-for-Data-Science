```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel("linechart.xlsx")
df
```

```python
plt.figure(figsize=(12,4))
plt.plot(df['Quarter'], df['Fridge'], color = "blue", label = "Fridge")
plt.plot(df['Quarter'], df['Dishwasher'], color = "red", label = "Dishwasher")
plt.plot(df['Quarter'], df['Washing Machine'], color = "green", label = "Washing Machien")
plt.title("Product Sales")
plt.ylabel("Revenue in USD")
plt.xlabel("Financial Quarter")
plt.legend()
plt.show()
```

```python
total_sales = df[["Fridge", "Dishwasher", "Washing Machine"]].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct="%1.2f%%", explode= (0.1,0,0), startangle= 140)
plt.show()
```
