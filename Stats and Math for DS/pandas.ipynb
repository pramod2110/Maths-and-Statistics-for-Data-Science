```python
df_wcity = pd.read_csv("/Users/pramodk/Desktop/Data Science Python/Maths-and-Statistics-for-Data-Science/Stats and Math for DS/weather_by_cities.csv")
df_wcity
```

```python
#1 simpler method is to use == sign for small dfs

df_wcity[df_wcity['city'] == 'new york']
```

```python
df_wcity[df_wcity['city'] == 'new york'].temperature.max()
```

```python
#Better way is to use groupby function

g = df_wcity.groupby("city")
g
```

```python
for city, data in g:
    print(city, "\n")
    print(data)
```

```python
#The data will be stored in grouops of city - essentially 3 new dfs inside g

g.get_group("mumbai")
```

```python
g.max()
```

```python
g.size()
```

```python
g.count()
```

```python
#group by can also be used with a function to build custom groups
def grouper(df, idx, col):
    if 80 <= df[col].loc[idx] <= 90:
        return '80-90'
    elif 50 <= df[col].loc[idx] <= 60:
        return '50-60'
    else:
        return 'others'
```

```python
g = df_wcity.groupby(lambda x: grouper(df_wcity, x, 'temperature'))
g
```

```python
for key, data in g:
    print("key", key)
    print(data)
    print("\n")
```

Data Concatenation and Merging

```python
india_weather = pd.DataFrame({
    "city": ['mumbai', 'delhi', 'banglore'],
    "temperature": [32, 45, 30],
    "humidity": [80, 60, 78]
})
```

```python
india_weather
```

```python
usa_weather = pd.DataFrame({
    "city": ['new york', 'chicago', 'orlando'],
    "temperature": [21, 14, 35],
    "humidity": [68, 65, 75]
})
```

```python
usa_weather
```

```python
df_merge = pd.concat([india_weather, usa_weather],  keys= ["india", "usa"])
```

```python
df_merge.loc["usa"]
```

```python
#another method is usign the merge functionx
df1 = pd.DataFrame({
    "city": ["chennai", "coimbatore", "trichy", "madurai"],
    "temparature": [40,25,35, 45]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chennai", "trichy", "kanyakumari"],
    "humidity": [60,59, 40]
})
df2
```

```python
pd.merge(df1,df2, on = "city")
```
