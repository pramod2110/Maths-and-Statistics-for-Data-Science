```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('bill_survey.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.City.nunique()
```

```python
sns.histplot(df.Bill, kde= True)
plt.show()
```

Let us plot for each sample (city)

```python
sns.histplot(df[df.City == 'Mumbai'], kde = True, bins = 20)
plt.show()
```

```python
sns.histplot(df[df.City == 'Surat'], kde = True, bins = 20)
plt.show()
```

```python
sample_means = []
for city, subdf in df.groupby('City'):
    sample_means.append(subdf.Bill.mean())
    
sample_means
```

```python
sns.histplot(sample_means, kde= True)
```
