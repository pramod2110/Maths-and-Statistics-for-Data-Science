```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/pramodk/Desktop/Data Science Python/Maths-and-Statistics-for-Data-Science/Stats and Math for DS/heights_NormalDist.csv')
df.head()
```

```python
df.shape
```

```python
sns.histplot(df.height, kde= True)
```

```python
mean = round(df.height.mean(), 2)
std = round(df.height.std(),2)

mean, std
```

```python
# Plot the histogram with KDE
sns.histplot(df['height'], kde=True)

# Add vertical lines for mean, and mean + std
plt.axvline(mean, color='black', linestyle='--', linewidth=1, label=f'Mean: {mean:.2f}')
plt.axvline(mean + std, color='r', linestyle='--', linewidth=1, label=f'Mean + 1 Std: {mean + std:.2f}')
plt.axvline(mean - std, color='r', linestyle='--', linewidth=1, label=f'Std: {std:.2f}')

plt.axvline(mean + 2*std, color='g', linestyle='--', linewidth=1, label=f'Mean + 2 Std: {mean + std:.2f}')
plt.axvline(mean - 2*std, color='g', linestyle='--', linewidth=1, label=f'Mean - 2 Std:: {std:.2f}')

plt.axvline(mean + 3*std, color='b', linestyle='--', linewidth=1, label=f'Mean + 3 Std: {mean + std:.2f}')
plt.axvline(mean - 3*std, color='b', linestyle='--', linewidth=1, label=f'Mean - 3 Std: {std:.2f}')

# Add legend
#plt.legend()

# Show the plot
plt.show()
```

Now let us check the spread of the data btwn each std

```python
left1 = mean - std
right1 = mean + std
left1, right1
```

```python
df[(df.height>left1) & (df.height <right1)]
```

```python
left2 = mean - 2*std
right2 = mean + 2*std
left2, right2
```

```python
df[(df.height>left2) & (df.height <right2)]
```

```python
left3 = mean - 3*std
right3 = mean + 3*std
left3, right3
```

```python
df[(df.height>left3) & (df.height <right3)]
```

As you can see, majority (99.9%) of the data is covered between std 3
